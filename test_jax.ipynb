```python
import whisper_s2t
```

```python
model = whisper_s2t.load_model(model_identifier="large-v3", backend='CTranslate2')
```

```python
files = ['./Acceleration_Due_To_Gravity.mp3']
lang_codes = ['ml']
tasks = ['translate']
initial_prompts = [None]
```

```python
%%time
out = model.transcribe_with_vad(files,
                                lang_codes=lang_codes,
                                tasks=tasks,
                                initial_prompts=initial_prompts,
                                batch_size=32)
```
